## 6.1 Formatierungsbeispiele der verantwortlichen Person

```python
# Verschiedene Formatierungsmuster der verantwortlichen Person sammeln
import re

# Nur Dokumente mit Inhalt
docs_with_content = df[df['document_content'].notna() & (df['content_length'] > 0)]

print(f"Insgesamt {len(docs_with_content)} Dokumente mit Inhalt\n")
print("="*80)
print("FORMATIERUNGSBEISPIELE DER VERANTWORTLICHEN PERSON AM ANFANG:\n")

# Zeige die ersten 20 EintrÃ¤ge und deren AnfÃ¤nge
sample_docs = docs_with_content.head(20)

for idx, (i, row) in enumerate(sample_docs.iterrows(), 1):
    content = str(row['document_content'])[:500]  # Erste 500 Zeichen
    
    # Zeige die erste(n) Zeile(n) / Absatz(e)
    first_lines = content.split('\n')[:3]  # Erste 3 Zeilen
    
    print(f"\n--- Beispiel {idx} ---")
    print(f"Titel: {row['name'][:60]}...")
    print(f"Referat: {row['ZustÃ¤ndiges Referat']}")
    print("\nAnfang des document_content:")
    for line in first_lines:
        if line.strip():
            print(f"  {line}")
    print()

print("="*80)
```

```python
# Extraction der Fraktion und verantwortlichen Person - Test mit ersten 50 Dokumenten
import re

def extract_fraktion(text):
    """Versucht, die Fraktion(en) aus dem Text zu extrahieren"""
    if not text:
        return []
    
    fraktionen = []
    
    # Pattern 1: "Fraktion [Name] des Stadtrates"
    matches = re.findall(r'Fraktion\s+([A-Za-z\-/\s\(\)]+?)(?:\s+des Stadtrates|[\n:])', text)
    fraktionen.extend([m.strip() for m in matches])
    
    # Pattern 2: "[Name]-Fraktion im Stadtrat" (z.B. "CSU-FW-Fraktion")
    matches = re.findall(r'([A-Za-z\-/]+?)\s*-\s*(?:Fraktion|Fraktion im Stadtrat)', text)
    fraktionen.extend([m.strip() for m in matches])
    
    # Pattern 3: Aus Email-Adresse extrahieren (z.B. csu-fw-fraktion@muenchen.de)
    matches = re.findall(r'([a-z\-]+)-fraktion@', text)
    if matches:
        fraktionen.extend([m.replace('-', ' ').title() for m in matches if m not in fraktionen])
    
    # Duplikate entfernen und bereinigen
    fraktionen = list(set(f for f in fraktionen if f and len(f) > 3))
    
    return fraktionen

def extract_person(text):
    """Versucht, die verantwortliche Person zu extrahieren"""
    if not text:
        return None
    
    # Pattern 1: "Herrn / OberbÃ¼rgermeister [Name]"
    match = re.search(r'(?:Herrn\s+)?(?:Ober)?bÃ¼rgermeister(?:in)?\s+([A-Z][a-z]+(?:\s+[A-Z][a-z]+)+)', text)
    if match:
        return match.group(1).strip()
    
    # Pattern 2: Name nach "Initiative:"
    match = re.search(r'Initiative:\s+([A-Z][a-z]+(?:\s+[A-Z][a-z]+)*)', text)
    if match:
        return match.group(1).strip()
    
    # Pattern 3: Kontaktperson in Adressblock (Name vor "Zimmer" oder "Rathaus")
    match = re.search(r'([A-Z][a-z]+(?:\s+[A-Z][a-z]+)*)\s*[|\n]\s*Rathaus', text)
    if match:
        return match.group(1).strip()
    
    return None

# Test mit ersten 50 Dokumenten
docs_sample = docs_with_content.head(50)

print("EXTRACTION TEST: Fraktion und verantwortliche Person")
print("="*90)

extraction_results = []

for idx, (i, row) in enumerate(docs_sample.iterrows(), 1):
    content = str(row['document_content'])
    
    fraktionen = extract_fraktion(content)
    person = extract_person(content)
    
    extraction_results.append({
        'antrag_nr': idx,
        'titel': row['name'][:50],
        'fraktionen': fraktionen,
        'person': person
    })
    
    print(f"\n[{idx}] {row['name'][:60]}...")
    print(f"    Fraktionen: {', '.join(fraktionen) if fraktionen else 'N/A'}")
    print(f"    Person: {person if person else 'N/A'}")

# Statistik
print("\n" + "="*90)
print("STATISTIK:")
success_fraktion = sum(1 for r in extraction_results if r['fraktionen'])
success_person = sum(1 for r in extraction_results if r['person'])

print(f"âœ“ Fraktion extrahiert: {success_fraktion}/{len(extraction_results)} ({100*success_fraktion/len(extraction_results):.1f}%)")
print(f"âœ“ Person extrahiert: {success_person}/{len(extraction_results)} ({100*success_person/len(extraction_results):.1f}%)")
print(f"âœ“ Beide extrahiert: {sum(1 for r in extraction_results if r['fraktionen'] and r['person'])}/{len(extraction_results)}")
```
